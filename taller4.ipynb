# Punto 1: Pre-Procesamiento

- `[5 pts]` Leer el archivo `dialogos.csv` usando `pandas`
- `[10 pts]` Crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números y sin palabras vacias
- `[7 pts]` Usando la nueva columna, ¿cuáles son las 10 palabras más usadas? 

```python
datos=pd.DataFrame(pd.read_csv("dialogos.csv"))
datos
```

```python
nltk.download('stopwords')
stopwords_sp = stopwords.words('spanish')
```

```python
datos['pre-procesado']=""
```

```python
cadena=""
for i in range(0, len(datos['Locución'])):
    dep=datos['Locución'][i] # Guardar el contenido de la celda en la variable dep
    dep=dep.lower() # Comverti a minuscula
    #dep=re.sub(r"[^a-zA-Z\s\d]", " ", dep)  
    dep=re.sub(r"[^a-zA-Záéíóúñ\s]", " ", dep) # Dejar solo letras, numeros, # y @
    dep=dep.split() # crear array con las palabras
    texto=[palabra for palabra in dep if palabra not in stopwords_sp] # Eliminar palabras vacias 
    Cadena=""
    for ele in texto:
        espacio=" "
        cadena += espacio
        cadena += ele       
    datos['pre-procesado'][i]=cadena
    cadena=""
```

```python
datos
```

```python
cadena=""
for i in range(0, len(datos['pre-procesado'])):
    cadena=cadena+" " + datos['pre-procesado'][i]
cadena
```

```python
cadena=cadena.split()
cadena
```

```python
import collections
```

```python
ocurrencias=collections.Counter(cadena)
```

```python
ocurrencias.most_common(10)
```
